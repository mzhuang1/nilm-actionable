```python
def find_transient_steady_power(df, steady_min=50, stead_max=200, transient_threshold=100 ):
    st = df[(df>50)&(df<200)]
    df2 = df.ix[argrelextrema(df.values, np.greater)[0]]
    tr = df2[df2>transient_threshold]
    return tr.mean(), st.mean()
```
